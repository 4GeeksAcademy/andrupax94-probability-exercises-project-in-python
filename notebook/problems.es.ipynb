## Ejercicio 1 

Dos dados se lanzan una vez y se observa el total obtenido. Usa una simulación para encontrar la probabilidad estimada de que la puntuación sea mayor a 7 o un número par. Una simulación es una repetición del mismo experimento multiples veces para observar su comportamiento:

- Ejecuta el experimento 1000 veces (lanza 2 dados 1000 veces, y suma el número de ambos dados).
- Lleva ua cuenta de los números y las veces que la suma fue mayor a 7 o un número par.
- Divide el número del paso 2 entre el número de iteraciones(1000).

```python
import random as rd

#inicializamos el contador
mayor7_o_par = 0
for i in range(1000):
    #lanzamos dos dados con randint
    d1 = rd.randint(1, 6)
    d2 = rd.randint(1, 6)
    
    #los sumamos
    suma = d1 + d2
    
    #vemos la probabilidad de que se cumplan ambas condiciones, que sea ayor a 7 y que sea par
    if suma > 7 or suma % 2 == 0:
        mayor7_o_par += 1

#construimos el resultado dividiendolo entre 1000, multiplicandolo por 100 para mover los decimales dos pasos hacia adelante, y redondeamos a 2 decimales        
resultado = f"{round((mayor7_o_par / 1000) * 100, 2)}%"
#imprimimos
resultado=f"La probabilidad de que la suma de los dados sea par o mayor que 7 es: {resultado}, aproximadammente"
resultado
```

## Ejercicio 2

Una caja contiene 10 bolas blancas, 20 bolas rojas y 30 bolas verdes. Si tomamos 5 bolas de la caja reemplazándolas (tomando una bola, anotando el color y luego regresandola a la caja). Queremos saber la probabilidad de:

1. Tomar 3 blancas y 2 rojas.
2. Toas todas del mismo color.

Ejecuta el experimento 1000 veces y calcula las probabilidades que mencionadas.

```python
ball_box = {}

# Crea la cajas con las bolas
for i in range(60):
    if i < 10:
        ball_box[i] = "White"
    elif (i > 9) and (i < 30):
        ball_box[i] = "Red"
    else:
        ball_box[i] = "Green"

#definimos la funcion con el parametro "n" de las iteraciones por defecto "1000"          
def experiment(n = 1000):
    #inicializamos los contadores
    tres_y_dos = 0
    mismo_color = 0 
    for i in range(n):
        
        #inicializamos los contadores de bolas por color
        bolas_blancas = 0
        bolas_rojas = 0
        bolas_verdes = 0
        for x in range(5):
            #numero aleatorio del total de bolas
            n_bola = rd.randint(0,59)
            #obtenemos el color
            bola = ball_box[n_bola]
            #y lo comparamos para sumar al contador que corresponda
            if bola == "White":
                bolas_blancas += 1
            elif bola == "Red":
                bolas_rojas += 1
            elif bola == "Green":
                bolas_verdes += 1
        #vemos si se cummple la primera condicion de 3 blancas y 2 rojas  
        if (bolas_blancas == 3) and (bolas_rojas == 2):
            tres_y_dos += 1
        # y luego vemos si se cumple la segunda condicion de todas del mismo color
        if (bolas_blancas == 5) or (bolas_rojas == 5) or (bolas_verdes == 5):
            mismo_color += 1
    #hacemos lo mismo que el ejercicio anterior y contruimos el resultado
    p_tres_y_dos = f"{round((tres_y_dos / n) * 100, 2)}%"
    p_mismo_color = f"{round((mismo_color / n) * 100, 2)}%"
    #lo devolvemos
    return f"""La probabilidad de que salgan tres blancas y dos rojas es:{p_tres_y_dos}, 
Mientras que la probabilidad de que las 5 sean del mismo color es:{p_mismo_color}, Aproxiadamente"""
                    
resultado=experiment(1000)
resultado       
```
