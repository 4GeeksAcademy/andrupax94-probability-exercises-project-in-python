## Ejecicio 1
1. Qué es combinatoria, permutación y como esto se relaciona con probabilidad.
    1. 1 Diferencia entre en combinatoria o permutación.

### Convinatoria:
    las convinatorias son una rama de las matermaticas que estudian cuantas formas o maneras distintas hay de: valga la redundancia "combinar elementos",Acomodarlos,Seleccionarlos. Se usan en la probabilidad porque muchas veces necesitamos saber: cuantos resultados posibles hay y cuales de esos resultados son favorables para aplicar las funcion de P= casos favorables / casos posibles.

### Permutaciones:
    Las permutaciones forman parte de las convinatorias pero con una diferencia importante "el orden si importa" se usan para saber cuantas formas hay de ordenar "x" elementos.

### Diferencias:
    El principal criterio que las diferencia es la importancia del orden al contar, mientras que una convinatoria el orden de los factores no afecta al conteo, en una permutacion si, a efectos practicos en una combinatoria esto seria igual (A,B,C==B,C,A) y ambos solo contarian 1, mientras que en una permutacion (A,B,C!=B,C,A) son diferentes por lo que contarian 2, si a una serie de elementos le aplicamos ambas contabilizaciones obtendremos mas numeros de casos en las permutaciones y menos numero de casos en las convinatorias 


## Ejericio 2
2.  Un entrenador dispone de 22 jugadores para formar un equipo de fútbol.
¿Cuántas alineaciones de 11 jugadores puede hacer?

```python
#como aca no tenemos en cuenta las posiciones de los jugadores el orden no importa por lo que aplicamos una convinatoria con math.comb
alineaciones = ma.comb(22, 11)
resultado = f"el numero de alineaciones posibles es: {alineaciones}"
resultado
```

## Ejericio 4
4. ¿De cuántas formas se pueden sentar tres personas en seis sillas?.

```python
#aqui si cambia la cosa pues el orden si importa entonses usamos math.perm para hacer una permutacion
formas_sentarse  = ma.perm(6, 3)
resultado = f"El numero de formas de sentarse es: {formas_sentarse}"
resultado
#como vimos el resultado de una permutacion arroja un mayor numero de casos si lo comparamos con un ejericio parecido como anterior
#permutacion:120 vs convinatoria:20
```

## Ejercicio 5
5. ¿Cuántas jugadas diferentes se pueden obtener si se sacan cinco cartas de una
baraja de 40 cartas?

```python
#Aca depende del juego, puede variar de si el orden importa o no: en la mayoria de juegos de cartas el orden no importa puedes hacer una jugada
#aunque saques una y luego otra, pero pero vamos a calcular ambos de igual modo
jugadas_comb = ma.comb(40, 5)
jugadas_perm = ma.perm(40, 5)
resultado = f"el numero de jugadas posibles si no importa el orden: {jugadas_comb}"
resultado2 = f"el numero de jugadas posibles si importa el orden: {jugadas_perm}"
print(resultado)
print(resultado2)
```

## Ejercicio 6
6. ¿De cuántas formas se pueden sentar cinco personas en una fila de butacas de
un cine?

```python
#Suponiendo que la fila es de 5 personas , es decir 5 asientos estariamos halando de una permutacion completa
#las permutaciones completas se pueden obtener solamente sacando el factorial del numero de personas
formas = ma.factorial(5)
resultado = f"formas de sentar a 5 personas en una fila: {formas}"
resultado
```
